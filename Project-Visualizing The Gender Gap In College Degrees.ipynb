```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees.head()
```

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 
             'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', 
            linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', 
            linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

# Comparing across all degree categories

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize = (16,20))
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

# Generate first column of line charts. STEM degrees.

for i in range(0,18,3):
    ind = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[ind]], linewidth = 3, c = cb_dark_blue)
    ax.plot(women_degrees["Year"], 100- women_degrees[stem_cats[ind]], linewidth = 3, c=cb_orange)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[ind])
    ax.tick_params(bottom = "off", top = "off", left = "off", right = "off", labelbottom = "off")
    ax.set_yticks((0,100)) # Setting y-axis labels
    ax.axhline(50, c = (171/255,171/255,171/255), alpha = 0.3) # Adding a horizontal line. alpha represent transparency of line
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if ind == 0:
        ax.text(2004,80,"Women")
        ax.text(2004,15,"Men")
    if ind == 5:
        ax.text(2004,20,"Women")
        ax.text(2004,90,"Men")
        ax.tick_params(labelbottom = "on") # appearing x-axis labels
    
# Generate second column of line charts. Liberal arts degrees.
for i in range(1,16,3):
    cat = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[cat]], linewidth = 3, c = cb_dark_blue)
    ax.plot(women_degrees["Year"], 100- women_degrees[lib_arts_cats[cat]], linewidth = 3, c=cb_orange)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat])
    ax.tick_params(bottom = "off", top = "off", left = "off", right = "off", labelbottom = "off")
    ax.set_yticks([0,100])
    ax.axhline(50,c = (171/255,171/255,171/255), alpha = 0.3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if cat == 0:
        ax.text(2004,74,"Women")
        ax.text(2004,20,"Men")
    if cat == 4:
        ax.tick_params(labelbottom = "on")
        
# Generate third column of line charts. Other degrees.
for i in range(2,18,3):
    inde = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees["Year"], women_degrees[other_cats[inde]], linewidth = 3, c = cb_dark_blue)
    ax.plot(women_degrees["Year"], 100- women_degrees[other_cats[inde]], linewidth = 3, c=cb_orange)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[inde])
    ax.tick_params(bottom = "off", top = "off", left = "off", right = "off", labelbottom = "off")
    ax.set_yticks([0,100])
    ax.axhline(50, c = (171/255,171/255,171/255))
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if inde == 0:
        ax.text(2004,90,"Women")
        ax.text(2004,7,"Men")
    if inde == 5:
        ax.text(2004,65,"Men")
        ax.text(2004,35,"Women")
        ax.tick_params(labelbottom = "on")
        
plt.show()       
```
